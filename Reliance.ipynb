```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
data = pd.read_csv("/content/drive/MyDrive/Collab/stock/Stocks/Indian_companies-2007:2020/RELIANCE.csv")
```

```
data.shape
```

```
data.tail()
```

```
data = data.iloc[0:3000,:]
```

```
import seaborn as sns
plt.figure(figsize=(30,15) )

sns.lineplot(x= data["Date"], y = data["Close"],color = "blue"  )
```

```
rolling_mean_ = data.Close.rolling(window=50).mean()
rolling_mean2_ = data.Close.rolling(window=200).mean()
```

```
plt.figure(figsize=(30,15) )
plt.plot(data.Date, data.Close)
plt.plot(data.Date, rolling_mean_, label='50 Day SMA', color='green')
plt.plot(data.Date, rolling_mean2_, label='200 Day SMA', color='red')
plt.show()
```

Here im using golden cross over means 200 ma crosses 50 ma
here we are representing 50ma with red color and 200 ma with green color.

If the 50ma crosses above to 200ma its a buy signal and vice versa

```
rolling_mean_ = datanew.Close.rolling(window=50).mean()
rolling_mean2_ = datanew.Close.rolling(window=200).mean()
```

```
plt.figure(figsize=(30,15) )
plt.plot(datanew.Date, datanew.Close)
plt.plot(datanew.Date, rolling_mean_, label='50 Day SMA', color='green')
plt.plot(datanew.Date, rolling_mean2_, label='200 Day SMA', color='red')   
plt.show()
```
